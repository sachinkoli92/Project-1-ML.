```python
titanic.drop(['Cabin'],axis=1,inplace=True)
```

```python
from sklearn.impute import SimpleImputer
si=SimpleImputer(strategy='median')
titanic['Age']=si.fit_transform(titanic['Age'].values.reshape(-1,1))
```

```python
titanic.dropna(inplace=True)
```

```python
sns.heatmap(titanic.isnull()) 
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
list=['Sex','Embarked']
for i in list:
    titanic[i]=le.fit_transform(titanic[i].astype(str))
```

```python
titanic['Sex']
```

```python
titanic['Embarked']
```

```python
titanic['Pclass']=le.fit_transform(titanic['Pclass'].astype('int'))
```

```python
titanic.drop(['Name','PassengerId','Ticket','Fare'],axis=1,inplace=True)
```

```python
titanic
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
x=titanic.drop(['Survived'],axis=1)
x
```

```python
y=titanic['Survived']
y
```

```python
lg=LogisticRegression()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=68)
model=[LogisticRegression(), GaussianNB() , SVC(), DecisionTreeClassifier()]
for m in model:
    m.fit(x_train,y_train)
    m.score(x_train,y_train)
    predm=m.predict(x_test)
    print("accuracy score of", m, "is", accuracy_score(y_test,predm))
    print("confusion matrix:", confusion_matrix(y_test,predm))
    print("classification report:", classification_report(y_test,predm))
    print("\n")
```

In the train split model Decision Tree Classifier algorithm is perfroming well.

# CROSS VALIDATION SCORE

```python
from sklearn.model_selection import cross_val_score
list=[GaussianNB(),SVC(),LogisticRegression(),DecisionTreeClassifier(),KNeighborsClassifier()]
for i in list:
    cvscore=cross_val_score(i,x,y,cv=8)
    print('The cross validation score for',i,'is',cvscore)
    print('The mean and standard deviations are:',cvscore.mean(),cvscore.std())
    print('\n')
```

In the cross validation model, the algorithm that is performing well is Logistic regression

```python
max_r_score=0
a=[GaussianNB(),SVC(),LogisticRegression(),DecisionTreeClassifier(),KNeighborsClassifier()]
for m in a:
    print('for the algorithm:',m)
    for i in range(33,101):
        x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=i,test_size=.20)
        m.fit(x_train,y_train)
        pred=m.predict(x_test)
        score=m.score(x_train,y_train)
        accuracy=accuracy_score(pred,y_test)
        print('accuracy score corresponding to random state',i,'is: ',accuracy)
        if accuracy>max_r_score:
            max_r_score=accuracy
            final=i
            
        print('\n')
        print('Max accuracy score corresponding to: ', final,'is:',max_r_score)
```

Using Hyperprameter tuning we have observed that in the train test split model Decision Tree Classifier is working well at Random State of 68, providing the accuracy score of 0.8820224719101124.

```python
dtc=DecisionTreeClassifier(criterion='gini')
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=68)
dtc.fit(x_train,y_train)
print(dtc.score(x_train,y_train))
pred=dtc.predict(x_test)
print("Accuracy score:", accuracy_score(y_test,pred))
print("Confusion matrix:",confusion_matrix(y_test,pred))
print('Classification report:', classification_report(y_test,pred))
```

```python
import joblib
joblib.dump(dtc,'titanic_dtc.obj')
dtc_from_joblib=joblib.load('titanic_dtc.obj')
dtc_from_joblib.predict(x_test)
```
