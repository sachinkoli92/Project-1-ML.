```python
df.drop(['Total phenols','Flavanoids','diluted wines'],axis=1,inplace=True)
```

```python
df.skew()
```

```python
#removing skewness using logtransform
import numpy as np
import warnings
warnings.filterwarnings('ignore')
for i in df.columns:
    if df.skew().loc[i]>0.55:
        df[i]=np.log1p(df[i])
```

```python
df.skew()
```

```python
from scipy.stats import zscore
```

```python
z=np.abs(zscore(df))
print(df.shape)
```

```python
z
```

```python
wine=df[(z<3).all(axis=1)]
```

```python
wine
```

```python
wine.skew()
```

```python
x=wine.iloc[:,0:-1]
x
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x=sc.fit_transform(x)
```

```python
x
```

```python
x=pd.DataFrame(data=x)
x
```

```python
y=wine.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
list=[GaussianNB(),SVC(),DecisionTreeClassifier(),KNeighborsClassifier(),RandomForestClassifier()]
for j in list:
    r_max=0
    print('for the algorithm:',j)
    for i in range(33,101):
        x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=i,test_size=.30)
        j.fit(x_train,y_train)
        pred=j.predict(x_test)
        score=j.score(x_train,y_train)
        accuracy=accuracy_score(pred,y_test)
        print('accuracy score corresponding to random state',i,'is:',accuracy)
        if accuracy>r_max:
            r_max=accuracy
            final=i
    
    print('\n')
    print('Max accuracy score corresponding to:', final, 'is: ',r_max)
    print('\n')
```

from above code we have got certain observations:
1.Max. accuracy score for support vector classifier corresponding to random state 34 is= 1.0
2.Max. accuracy score for GaussianNB corresponding to random state 34 is= 1.0
3.Max. accuracy score for DecisionTreeClassifier to random state 68 is = 1.0
4.Max. accuracy score for KNeighborsClassifier to random state 34 is = 1.0
5.Max. accuracy score for RandomForestClassifier to random state 43 is = 1.0

```python
from sklearn.model_selection import cross_val_score
list=[GaussianNB(),SVC(),DecisionTreeClassifier(),KNeighborsClassifier(),RandomForestClassifier()]
for i in list:
    cvscore=cross_val_score(i,x,y,cv=9)
    print('The cross validation score for', i, 'is: ',cvscore)
    print('The mean and standard Deviations are: ', cvscore.mean(),cvscore.std())
    print('\n')
```

We can observe that in cross validation score model SVC is performing well with average accuracy 0.976608

```python
#working on the best model-SVC
svc=SVC (kernel='rbf')
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=34)
svc.fit(x_train,y_train)
print(svc.score(x_train,y_train))
pred=svc.predict(x_test)
print("Accuracy score:",accuracy_score(y_test,pred))
print("Confusion matrix:",confusion_matrix(y_test,pred))
print('Classification report:',classification_report(y_test,pred))
```

```python
import joblib
joblib.dump(svc,'wine_svc.obj')
svc_load=joblib.load('wine_svc.obj')
svc_load.predict(x_test)
```
